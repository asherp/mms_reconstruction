```python
from plotly.offline import iplot, plot, init_notebook_mode
import plotly.graph_objs as go
import pandas as pd
```

```python
init_notebook_mode(connected = True)
```

```python
def plot_array(array, title = None, index = None, names = None):
    df = pd.DataFrame(array, index = index)
    traces = []
    for i, c_ in enumerate(df.columns):
        name = None
        if names is not None:
            if type(names) == str:
                name = names.format(c_)
            else:
                try:
                    name = names[i]
                except:
                    pass
        traces.append(go.Scatter(x = df.index, y = df[c_], name = name))
    return go.Figure(traces, layout = dict(title = title))
```
